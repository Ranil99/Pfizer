```python
pf_vac['user_created'] = pd.to_datetime(pf_vac['user_created'])
pf_vac['date'] = pd.to_datetime(pf_vac['date'])
```

```python
pf_vac.info()
```

Analysis Begins

```python
pf_vac['date'] = pf_vac['date'].dt.date
```

```python
date = pf_vac.groupby('date').count().reset_index()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,8))
g = sns.lineplot(x='date', y='id', data=date, color='orange', markers='o')
sns.despine(left=True)
g.set_title('Tweets per Day', fontsize=25)
g.set_xlabel('Date', fontsize=20)
g.set_ylabel('Number of Tweets', fontsize=20)
```

```python
htags = pd.DataFrame(pf_vac['hashtags'].value_counts()).reset_index()
htags = htags[htags['index'] != 'No HashTag']
htags = htags.rename(columns={'index': 'Hashtags', 'hashtags': 'counts'})[:10]
```

```python
sns.set_style('white')
plt.figure(figsize=(12,8))
h = sns.barplot(x='Hashtags', y='counts', data=htags, palette='magma_r')
sns.despine(left=True)
h.set_title('Top 10 Hastag Most Used', fontsize=25)
h.set_xlabel('#Hashtags', fontsize=20)
h.set_ylabel('Number of Tags', fontsize=20)
plt.xticks(rotation=90)
```

```python
top_10 = htags['Hashtags']
retweet = pf_vac[pf_vac['hashtags'].isin(top_10)]

sns.set_style('white')
plt.figure(figsize=(12,8))
k = sns.barplot(x='hashtags', y='retweets', data=retweet, color='red')
sns.despine(left=True)
k.set_title('Top 10 Hashtag Most Retweeted', fontsize=25)
k.set_xlabel('#Hashtags', fontsize=20)
k.set_ylabel('Number of Retweets', fontsize=20)
plt.xticks(rotation=90)
```

```python
pf_vac['user_created'] = pf_vac['user_created'].dt.date
```

```python
sns.set_style('dark')
plt.figure(figsize=(12,12))
b = sns.scatterplot(x='user_created', y='id', data=pf_vac, palette='magma', hue='user_verified', size='user_verified', sizes= (50,200), size_order=[True, False])
sns.despine(left=True)
b.set_title('Verified Account Check For Users!', fontsize=25)
b.set_xlabel('Date n which Accounts were created', fontsize=20)
b.set_ylabel('Number of Accounts', fontsize=20)
b.set_xlim(pf_vac['user_created'].min(), pf_vac['user_created'].max())
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
top_10_pf = pf_vac[pf_vac['hashtags'].isin(top_10)]
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,8))
j = sns.scatterplot(x='user_friends', y='user_followers', data=top_10_pf, hue='hashtags', palette='inferno_r', s=70)
j.set_xlabel('User Friends', fontsize=15)
j.set_ylabel('User Followers', fontsize=15)
j.set_title('User Friends and Followers According to Hashtag Used', fontsize=20)
sns.despine(left=True)
plt.xlim(0,3000)
plt.ylim(top_10_pf['user_friends'].min(),3000)
plt.legend(bbox_to_anchor=(1.05,1), loc=2, borderaxespad=0.)
```

```python
top_source = top_10_pf['source'].value_counts()
top_source = pd.DataFrame(top_source).reset_index()
```

```python
sns.set_style('white')
plt.figure(figsize=(14,8))
o = sns.barplot(data=top_source, x='index', y='source', alpha=0.7, palette='RdBu')
sns.despine(left=True)
o.set_xlabel('Source', fontsize=15)
o.set_ylabel('Counts', fontsize=15)
o.set_title('Sources Used for Tweets', fontsize=20)
```
